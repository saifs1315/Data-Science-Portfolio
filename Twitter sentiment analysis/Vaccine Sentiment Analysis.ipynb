# PFIZER VACINE TWEETS SENTIMENT ANALYSIS

The aim of this project is to use a dataset containing tweets about Pfizer & BioNTech vaccine and perform text preprocessing, sentiment analysis and creating a classification model that can classify the sentiment of the tweets in the dataset. The dataset was downloaded from https://www.kaggle.com/datasets/gpreda/pfizer-vaccine-tweets?resource=download. 

The first step involves importing the required libraries and the respective modules:

1) Numpy and Panda are used for the mathematical operations and dataframe manipulation respectively

2) Matplotlib and seaborn are used for visual representation in the form of graphs

3) NLTK is used for text processing

4) WordCloud is used to create informative wordclouds containing frquently occuring words

5) TextBlob is used to process text and obtain sentiment of text

6) Sklearn is used for feature extraction, model creation and performance metrics

```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
# ignore python notebook warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('vaccination_tweets.csv')
```

Dropping the remaining columns and preserving the text column, provides us the dataset we will be performing our project objectives on.

```python
text_df = df.drop(labels=['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date', 'hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'],axis=1)
```

```python
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
```

```python
def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www|S+https\S+",'',text,flags=re.MULTILINE)
    text = re.sub(r"\@w+|\#",'',text)
    text = re.sub(r"[^\w\s]",'',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
text_df.text = text_df['text'].apply(data_processing)
```

```python
text_df = text_df.drop_duplicates('text')
```

```python
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

```python
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
```

```python
text_df.info()
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head()
```

```python
def sentiment(label):
    if label<0:
        return "Negative"
    elif label ==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

```python
text_df.head()
```

```python
fig = plt.figure(figsize=(5,5))
sns.countplot(x='sentiment',data=text_df)
plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
colors = ('Yellowgreen',"gold","red")
tags = text_df['sentiment'].value_counts()
explode = (0.1,0.1,0.1)
wp = {'linewidth':2,'edgecolor':"black"}
tags.plot(kind='pie',colors=colors,explode=explode,shadow=True, startangle=90,label='',autopct="%1.1f%%",wedgeprops=wp)
plt.title('Distribution of sentiments')
```

```python
pos_tweets = text_df[text_df['sentiment']=='Positive']
pos_tweets = pos_tweets.sort_values('polarity',ascending=False)
pos_tweets.head()
```

```python
text = ' '.join([word for word in pos_tweets['text']])
plt.figure(figsize=(20,15),facecolor='None')
wordcloud = WordCloud(max_words=500,width=1600,height=800).generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.title("Most frequent words in postive tweets",fontsize=19)
plt.show()
```

```python
neg_tweets = text_df[text_df['sentiment']=='Negative']
neg_tweets.sort_values('polarity',ascending=False)
neg_tweets.head()
```

```python
neg_text = ' '.join([word for word in neg_tweets['text']])
plt.figure(figsize=(20,15),facecolor='None')
wordcloud = WordCloud(max_words=500,width=1600,height=800).generate(neg_text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.title("Most frequent words in negative tweets",fontsize=19)
plt.show()
```

```python
neutral_tweets = text_df[text_df['sentiment']=='Neutral']
neutral_tweets.sort_values('polarity',ascending=False)
neutral_tweets.head()
```

```python
neutral_text = ' '.join([word for word in neutral_tweets['text']])
plt.figure(figsize=(20,15),facecolor='None')
wordcloud = WordCloud(max_words=500,width=1600,height=800).generate(neg_text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.title("Most frequent words in Neutral tweets",fontsize=19)
plt.show()
```

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

```python
feature_names = vect.get_feature_names()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features:\n {}".format(feature_names[:20]))
```

```python
X = text_df['text']
Y = text_df['sentiment']
X = vect.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print("Size of X train: {}".format(x_train.shape))
print("Size of X test: {}".format(x_test.shape))
print("Size of y train: {}".format(y_train.shape))
print("Size of y test: {}".format(y_test.shape))
```

```python
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred,y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test,logreg_pred))
print("\n")
print(classification_report(y_test,logreg_pred))
```

```python
style.use('classic')
cm = confusion_matrix(y_test,logreg_pred,labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=logreg.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
grid = GridSearchCV(logreg,param_grid)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(x_test)
```

```python
logreg_acc = accuracy_score(y_pred,y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test,y_pred))
print("\n")
print(classification_report(y_test,y_pred))
```
